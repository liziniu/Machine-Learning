```python
from sklearn import datasets
from sklearn.svm import SVR
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boston = datasets.load_boston()
data, target = boston.data, boston.target
print(data.shape, target.shape)
```

```python
plt.plot(target, color='blue')
plt.title('boston home price')
```

```python
# 3
# 3-order poly kernel
svr = SVR(kernel='poly', degree=3, max_iter=100)
svr.fit(data, target)
poly_y = svr.predict(data)

## 
## it is unstable for poly kernel
plt.plot(target, color='blue')
plt.plot(poly_y, color='red')
```

```python
## 
## rbf kernel
svr = SVR(kernel='rbf', max_iter=100)
svr.fit(data, target)
rbf_y = svr.predict(data)

## 
## it is smooth for rbf kernel
plt.plot(target, color='blue')
plt.plot(rbf_y, color='green')
plt.ylim([-100, 120])
```

```python
## sigmoid kernel
svr = SVR(kernel='sigmoid', max_iter=100)
svr.fit(data, target)
rbf_y = svr.predict(data)

## it is also smooth for sigmoid kernel
plt.plot(target, color='blue')
plt.plot(rbf_y, color='yellow')
plt.ylim([-100, 120])
```

```python
## linear kernel
svr = SVR(kernel='linear', max_iter=100)
svr.fit(data, target)
rbf_y = svr.predict(data)

## 
## it is o
plt.plot(target, color='blue')
plt.plot(rbf_y, color='gray')
plt.ylim([-100, 120])
```
